```python
import numpy as np; np.random.seed(0)
import seaborn as sns; sns.set()
uniform_data = np.random.rand(10, 12)
ax = sns.heatmap(uniform_data, cmap=sns.color_palette(["#B8C8F9", "#9BAAD5", "#6F7D9F", "#44516A", "#354155", "#4D455B", "#875F7A", "#C17A99", "#E78CAE"]))
```

```python
from numpy import arange
x = arange(25).reshape(5, 5)
cmap = sns.diverging_palette(338, 255, s=80, l=73, center= "dark", as_cmap=True) 
ax = sns.heatmap(x, cmap=cmap)
```
